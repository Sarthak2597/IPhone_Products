# Cleaning the dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("IPhone.csv")
df
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df
```

```python
df.dtypes
```

```python
df[["Total Rating","No of Reviews"]] = df["No_of_Rating"].str.split("&",1,expand = True)
```

```python
df = df.drop("No_of_Rating", axis = 1)
```

```python
df
```

```python
df[["ROM","Specs"]] = df["Specifications"].str.split("ROM",1,expand = True)
```

```python
df
```

```python
df["Total Rating"] = df["Total Rating"].str.split(" ").str[0] #No. of ratings
```

```python
df
```

```python
df["No of Reviews"] = df["No of Reviews"].str.split(" ").str[0]
```

```python
df
```

```python
df = df.drop("Specifications", axis = 1)
```

```python
df
```

```python
df.to_csv("Cleaned_IPhone_dataset.csv")
```

# Store Dataset in MongoDB

```python
import pymongo
```

```python
def dftomongodb(dataset):
    connect = pymongo.MongoClient(host = "localhost", port = 27017)
    database = connect["FlipkartDB"]
    collection = database["IPhone"]
    
    for (row,rs) in dataset.iterrows():
        Product_Name = rs[0]
        Sale_Price = rs[1]
        Actual_Price = rs[2]
        Rating = rs[3]
        Offer = rs[4]
        Total_Rating = rs[5]
        No_of_Reviews = rs[6]
        ROM = rs[7]
        Specs = rs[8]
        
        d = {"Product_Name": Product_Name, "Actual_Price": Actual_Price,
            "Sale_Price": Sale_Price, "Offer": Offer, "Rating": Rating,
            "Total_Ratings": Total_Rating, "No_of_Reviews": No_of_Reviews,
            "Space": ROM, "Specifications": Specs}
        collection.insert_one(d)
        
    print("Done")
```

```python
dftomongodb(df)
```
