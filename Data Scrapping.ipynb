# Connecting to the desired Web-Page

```python
#Downloading important libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
#Establishing a connection with the url (Flipkart page)
urls = []
for var in range(1,16):
    url = ["https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=%i"%(var)]
    urls.append(url[0])
```

# Extracting information

```python
#Extracting data from the url for one product
def connect(ur):
    uClient = urlopen(ur)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    container = page_soup.find_all("div",{"class": "_2kHMtA"})
    return container
```

```python
con = connect(urls[0])
print(soup.prettify(con[1]))
```

```python
for i in range(0,15):
    con = connect(urls[i])
    for container in con:
        product_name = container.div.img["alt"]
        print(product_name)
```

```python
#Identifying classes and tags
for container in con:
    s_price = container.findAll("div",{"class":"_30jeq3 _1_WHN1"})
    actual_price = container.findAll("div", {"class":"_3I9_wc _27UcVY"})
    rating = container.findAll("div",{"class":"_3LWZlK"})
    p_rating = container.findAll("span", {"class":"_2_R_DZ"})
    offer = container.findAll("div", {"class": "_3Ay6Sb"})
    specs = container.findAll("ul",{"class":"_1xgFaf"})
    print(offer)
```

```python
def dataframe(containers):
    data = {"Product_name":[], "Sale_Price":[], "Actual_Price":[], "Rating":[], "No_of_Rating":[], "Offer":[], "Specifications":[]}
    for i in range(0,15):
        con = connect(containers[i])
        for container in con:
            product_name = container.div.img["alt"]
            data["Product_name"].append(product_name)
            
            s_price = container.findAll("div", {"class":"_30jeq3 _1_WHN1"})
            s_p = s_price[0].text
            data["Sale_Price"].append(s_p)
            
            try:
                actual_price = container.findAll("div", {"class":"_3I9_wc _27UcVY"})
                a_p = actual_price[0].text
                data["Actual_Price"].append(a_p)
            except IndexError:
                print("Same as sale price")
                data["Actual_Price"].append("Same Price")
                
            try:
                rating = container.findAll("div",{"class": "_3LWZlK"})
                r = rating[0].text
                data["Rating"].append(r)
            except IndexError:
                data["Rating"].append("No rating")
                
                
            try:
                n_rating = container.findAll("span",{"class": "_2_R_DZ"})
                n = n_rating[0].text
                data["No_of_Rating"].append(n)
            except IndexError:
                data["No_of_Rating"].append("No rating")
                
                
            try:
                offer = container.findAll("div",{"class": "_3Ay6Sb"})
                off = offer[0].text
                data["Offer"].append(off)
            except IndexError:
                data["Offer"].append("No offer")
                
                
            try:
                specs = container.findAll("ul",{"class": "_1xgFaf"})
                s = specs[0].text
                data["Specifications"].append(s)
            except IndexError:
                data["Specifications"].append("No specifications")
                
                
        print("done ", i)
            
    return data
                
            
```

```python
dataset = dataframe(urls)
```
